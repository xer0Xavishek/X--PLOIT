```python
class Student:
    def __init__(self, name, cgpa, credits=9, department="CSE"):
        self.name = name
        self.cgpa = cgpa
        self.credits = credits
        self.department = department

    def checkScholarshipEligibility(self):
        if self.cgpa >= 3.7:
            print(f"{self.name} is eligible for Merit-based scholarship.")
        elif self.cgpa >= 3.5 and self.credits > 10:
            print(f"{self.name} is eligible for Scholarship.")
        elif self.cgpa >= 3.5:
            print(f"{self.name} is eligible for Need-based scholarship.")
        else:
            print(f"{self.name} is not eligible for scholarship.")

    def showDetails(self):
        print(f"Name: {self.name}")
        print(f"Department: {self.department}")
        print(f"CGPA: {self.cgpa}")
        print(f"Number of Credits: {self.credits}")
        if self.cgpa >= 3.7:
            print("Scholarship Status: Merit-based scholarship")
        elif self.cgpa >= 3.5 and self.credits > 10:
            print("Scholarship Status: Scholarship")
        elif self.cgpa >= 3.5:
            print("Scholarship Status: Need-based scholarship")
        else:
            print("Scholarship Status: No scholarship")


print('--------------------------')
std1 = Student("Alif", 3.99, 12)
print('--------------------------')
std1.checkScholarshipEligibility()
print('--------------------------')
std1.showDetails()
print('--------------------------')
std2 = Student("Mim", 3.4)
std3 = Student("Henry", 3.5, 15,"BBA")
print('--------------------------')
std2.checkScholarshipEligibility()
print('--------------------------')
std3.checkScholarshipEligibility()
print('--------------------------')
std2.showDetails()
print('--------------------------')
std3.showDetails()
print('--------------------------')
std4 = Student("Bob", 4.0, 6, "CSE")
print('--------------------------')
std4.checkScholarshipEligibility()
print('--------------------------')
std4.showDetails()
```

```python
class Foodie:
    menu = {'Chicken Lollipop': 15, 'Beef Nugget': 20, 'Americano': 180, 'Red Velvet': 150, 'Prawn Tempura': 80, 'Saute Veg': 200}
    tips = 0
    
    def __init__(self, name):
        self.name = name
        self.cart = []
        self.total_spent = 0
    
    def order(self, *args):
        for item in args:
            food, quantity = item.split('-')
            if food.strip() in self.menu:
                self.cart.append(food.strip())
                price = self.menu[food.strip()] * int(quantity)
                self.total_spent += price
                print(f"Ordered - {food.strip()}, quantity - {quantity}, price (per Unit) - {self.menu[food.strip()]}.")
                print(f"Total price - {price}")
            else:
                print(f"{food.strip()} is not available in the menu.")
    
    def pay_tips(self, amount=0):
        self.tips += amount
        if amount > 0:
            print(f"Gives {amount}/- tips to the waiter.")
        else:
            print("No tips to the waiter.")
        self.total_spent += amount
    
    def show_orders(self):
        print(f"{self.name} has {len(self.cart)} item(s) in the cart.")
        if len(self.cart) > 0:
            print("Items: ", end="")
            print(*self.cart, sep=", ", end="")
            print()
        return f"Total spent: {self.total_spent}."

# Driver Code
menu = {'Chicken Lollipop': 15, 'Beef Nugget': 20, 'Americano': 180, 'Red Velvet': 150, 'Prawn Tempura': 80, 'Saute Veg': 200}
f1 = Foodie('Frodo')
print(f1.show_orders())
print('1----------------------')
f1.order('Chicken Lollipop-3','Beef Nugget-6','Americano-1')
print('2----------------------')
print(f1.show_orders())
print('3----------------------')
f1.order('Red Velvet-1')
print('4----------------------')
f1.pay_tips(20)
print('5----------------------')
print(f1.show_orders())

f2 = Foodie('Bilbo')
print('6----------------------')
f2.order('Prawn Tempura-6','Saute Veg-1')
print('7----------------------')
f2.pay_tips()
print('8----------------------')
print(f2.show_orders())
```

```python
class Department:
    def __init__(self, name='ChE Department', sections=5):
        self.name = name
        self.sections = sections
        self.students = 0
        self.average_students = 0
        print(f"The {self.name} has {self.sections} sections.")

    def add_students(self, *args):
        if len(args) == self.sections:
            self.students = sum(args)
            self.average_students = self.students / self.sections
            print(f"The {self.name} has an average of {self.average_students:.2f} students in each section.")
        else:
            print(f"The {self.name} doesn't have {len(args)} sections.")

    def merge_Department(self, *args):
        total_students = self.students
        total_sections = self.sections
        for department in args:
            total_students += department.students
            total_sections += department.sections
            print(f"{department.name} is merged to {self.name}.")
        self.students = total_students
        self.sections = total_sections
        self.average_students = self.students / self.sections
        return f"Now the {self.name} has an average of {self.average_students:.1f} students in each section."
    
    
d1 = Department()
print('1-----------------------------')
d2 = Department('MME Department')
print('2-----------------------------')
d3 = Department('NCE Department', 8)
print('3-----------------------------')
d1.add_students(12, 23, 12, 34, 21)
print('4-----------------------------')
d2.add_students(40, 30, 21)
print('5-----------------------------')
d3.add_students(12, 34, 41, 17, 30, 22, 32, 51)
print('6-----------------------------')
mega = Department('Engineering Department', 10)
print('7-----------------------------')
mega.add_students(21,30,40,36,10,32,27,51,45,15)
print('8-----------------------------')
print(mega.merge_Department(d1, d2))
print('9-----------------------------')
print(mega.merge_Department(d3))

```

```python
#task 3
class Department:
  def __init__(self,name="ChE Department",count=5):
    self.name = name
    self.secCount = count
    self.avgStudent = 0
    print(f"The {name} has {count} sections.")
  def add_students(self,*stds):
    if len(stds) < 5 :
      print(f"The {self.name} doesn't have {len(stds)} sections.")
    else :
      totalStu = 0.0
      totalsec = 0
      for i in stds:
        totalsec += 1
        totalStu += i
      self.secCount = totalsec
      self.avgStudent = round(totalStu/totalsec,2)
      print(f"The {self.name} has an average of {self.avgStudent} students in each section.")
  def merge_Department(self,*sec):
    totalStu = self.secCount * self.avgStudent
    for i in sec:
      totalStu += i.secCount * i.avgStudent
      self.avgStudent = round(totalStu/self.secCount,2)
      print(f"{i.name} is merged to {self.name}.")
    return(f"Now the {self.name} has an average of {self.avgStudent} students in each section.")
d1 = Department()
print('1-----------------------------')
d2 = Department('MME Department')
print('2-----------------------------')
d3 = Department('NCE Department', 8)
print('3-----------------------------')
d1.add_students(12, 23, 12, 34, 21)
print('4-----------------------------')
d2.add_students(40, 30, 21)
print('5-----------------------------')
d3.add_students(12, 34, 41, 17, 30, 22, 32, 51)
print('6-----------------------------')
mega = Department('Engineering Department', 10)
print('7-----------------------------')
mega.add_students(21,30,40,36,10,32,27,51,45,15)
print('8-----------------------------')
print(mega.merge_Department(d1, d2))
print('9-----------------------------')
print(mega.merge_Department(d3))
```

```python
class Shopidify:
    def __init__(self, name='Guest'):
        self.name = name
        self.cart = {}
        self.history = []
        print(f"Welcome {self.name} to Shopidify")

    def add_to_cart(self, *args):
        if isinstance(args[0], list):
            items = args[0]
            for i in range(0, len(items), 2):
                item = items[i]
                quantity = items[i+1]
                self.cart[item] = self.cart.get(item, 0) + quantity
        else:
            item = args[0]
            quantity = 1 if len(args) == 1 else args[1]
            self.cart[item] = self.cart.get(item, 0) + quantity

    def display_cart(self):
        print(f"Items in the cart for {self.name}:")
        for item, quantity in self.cart.items():
            print(f"- {item}: {quantity}x")

    def checkout(self):
        self.history.append(self.cart)
        self.cart = {}
        print(f"Checkout completed for {self.name}")

    def display_history(self):
        print(f"Purchase history for {self.name}:")
        for i, transaction in enumerate(self.history, 1):
            print(f"Transaction {i}:")
            for item, quantity in transaction.items():
                print(f"- {item}: {quantity}x")
                
                
# Test the Shopidify class
guest_account = Shopidify()
print("1xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account = Shopidify("John")
print("2xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.add_to_cart("Air Jordan", 2)
guest_account.add_to_cart("Luffy Action Figure")
guest_account.display_cart()
print("3xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.add_to_cart(["Chocolate Chip Cookies", 3,"Goku Action Figure",2,"Dumbbells-5kg",2])
john_account.display_cart()
print("4xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.add_to_cart("Air Jordan")
guest_account.display_cart()
print("5xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.checkout()
print("6xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.display_history()
print("7xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.checkout()
print("8xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.display_history()
print("9xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")                
                
```

```python
class Author:
    def __init__(self, name=None):
        self.name = name
        self.books = {}

    def addBook(self, title, genre):
        if self.name is None:
            print("A book can not be added without author name")
            return
        if genre not in self.books:
            self.books[genre] = [title]
        else:
            self.books[genre].append(title)

    def setName(self, name):
        self.name = name

    def printDetail(self):
        if self.name is None:
            print("Author name is not set.")
            return
        print(f"Number of Book(s): {sum(len(books) for books in self.books.values())}")
        print(f"Author Name: {self.name}")
        for genre, titles in self.books.items():
            print(f"{genre}: {', '.join(titles)}")


# Driver code
a1 = Author()
print("=================================")
a1.addBook("Ice", "Science Fiction")
print("=================================")
a1.setName("Anna Kavan")
a1.addBook("Ice", "Science Fiction")
a1.printDetail()
print("=================================")
a2 = Author("Humayun Ahmed")
a2.addBook("Onnobhubon", "Science Fiction")
a2.addBook("Megher Upor Bari", "Horror")
print("=================================")
a2.printDetail()
a2.addBook("Ireena", "Science Fiction")
print("=================================")
a2.printDetail()
print("=================================")
```
