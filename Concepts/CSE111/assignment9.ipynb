```python
class NikeBangladesh:
    branches_opened = []

    def __init__(self, location):
        self.location = location
        self.inventory = {"Air Jordan": 0, "Cortez": 0, "Zoom Kobe": 0}
        self.sold = 0
        NikeBangladesh.branches_opened.append(self)

    def details(self):
        print(f"Nike {self.location} outlet:")
        print("Products Currently Stocked:")
        print(self.inventory)
        print(f"Sold: {self.sold}")

    @classmethod
    def status(cls):
        print("Nike Bangladesh Status:")
        print(f"Branches Opened: {[branch.location for branch in cls.branches_opened]}")
        total_inventory = {"Air Jordan": 0, "Cortez": 0, "Zoom Kobe": 0}
        total_sold = 0
        for branch in cls.branches_opened:
            for product, quantity in branch.inventory.items():
                total_inventory[product] += quantity
            total_sold += branch.sold
        print("Currently Stocked")
        print(total_inventory)
        print(f"Sold: {total_sold}")

    def restockProducts(self, restock_dict):
        for product, quantity in restock_dict.items():
            self.inventory[product] += quantity

    def productSold(self, sold_dict):
        for product, quantity in sold_dict.items():
            self.inventory[product] -= quantity
            self.sold += quantity


print("xxxxxxxxxxxxxx1xxxxxxxxxxxxxxxx")
NikeBangladesh.status()
dhaka = NikeBangladesh("Dhaka Banani")
chittagong = NikeBangladesh("Chittagong GEC")
print("xxxxxxxxxxxxxx2xxxxxxxxxxxxxxxx")
dhaka.details()
print("xxxxxxxxxxxxxx3xxxxxxxxxxxxxxxx")
chittagong.details()
print("xxxxxxxxxxxxxx4xxxxxxxxxxxxxxxx")
dhaka.restockProducts({"Air Jordan": 1200, "Cortez": 200, "Zoom Kobe": 200})
chittagong.restockProducts({"Air Jordan": 1000, "Cortez": 250, "Zoom Kobe": 100})
print("xxxxxxxxxxxxxx5xxxxxxxxxxxxxxxx")
NikeBangladesh.status()
print("xxxxxxxxxxxxxx6xxxxxxxxxxxxxxxx")
dhaka.productSold({"Air Jordan": 760, "Cortez": 90})
chittagong.productSold({"Air Jordan": 520, "Zoom Kobe": 70})
print("xxxxxxxxxxxxxx7xxxxxxxxxxxxxxxx")
NikeBangladesh.status()
```

```python
class Player:
    total = 0
    enlisted_players = []

    def __init__(self, name="Neymar", number=10, team="Brazil"):
        self.name = name
        self.number = number
        self.team = team
        if self.name:
            Player.total += 1
            Player.enlisted_players.append(self.name)

    def set_name(self, name):
        self.name = name

    def set_number(self, number):
        self.number = number

    def set_team(self, team):
        self.team = team

    def player_detail(self):
        return f"Player Name: {self.name}\nJersey Number: {self.number}\nCountry: {self.team}"

    @classmethod
    def info(cls):
        print("Total number of players:", cls.total)
        print("Players enlisted so far:", ', '.join(cls.enlisted_players))
        print("---------------------------")


print("Total number of players:", Player.total)
print("---------------------------")
p1 = Player()
p1.set_name("Neymar")
p1.set_team("Brazil")
print(p1.player_detail())
print('========================')
Player.info()
print("---------------------------")
p2 = Player("Ronaldo")
p2.set_number(7)
p2.set_team("Portugal")
print(p2.player_detail())
print('========================')
Player.info()
print("---------------------------")
p3 = Player("Messi")
p3.set_team("Argentina")
print(p3.player_detail())
print('========================')
Player.info()
print("---------------------------")
p4 = Player("Mbappe", 10, "France")
print(p4.player_detail())
print('========================')
Player.info()
```

```python
class SultansDine:
    total_branches = 0
    total_sell = 0
    branches = {}

    def __init__(self, name):
        self.name = name
        self.sell = 0
        SultansDine.total_branches += 1
        SultansDine.branches[name] = self

    def sellQuantity(self, amount):
        if amount < 10:
            self.sell += amount * 300
        elif amount < 20:
            self.sell += amount * 350
        else:
            self.sell += amount * 400
        SultansDine.total_sell += self.sell

    @classmethod
    def details(cls):
        print(f"Total Number of branch(s): {cls.total_branches}")
        print(f"Total Sell: {cls.total_sell} Taka")
        print('#################################')

    def branchInformation(self):
        print(f"Branch Name: {self.name}")
        print(f"Branch Sell: {self.sell} Taka")
        print("-----------------------------------------")
        print(f"Total Number of branch(s): {SultansDine.total_branches}")
        print(f"Total Sell: {SultansDine.total_sell} Taka")
        for branch_name, branch in SultansDine.branches.items():
            total_percentage = (branch.sell / SultansDine.total_sell) * 100 if SultansDine.total_sell > 0 else 0
            print(f"Branch Name: {branch_name}, Branch Sell: {branch.sell} Taka")
            print(f"Branch consists of total sell's: {total_percentage:.2f}%")
        print('================================')
SultansDine.details()
print('########################')
dhanmondi = SultansDine('Dhanmondi')
dhanmondi.sellQuantity(25)
dhanmondi.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```
