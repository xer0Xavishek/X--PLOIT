```python
class PlayerEarning:
    
    def __init__(self,name):
        self.name=name
        self.earning=0
        self.bonus=0    
    
    def calculateTotal(self,earn,goals=0):
        self.earning=earn
        if goals!=0: 
            if goals>30 :
                self.bonus= int(5/100*earn+10000)

            else:
                self.bonus=int(5/100*earn)
                
    def printDetails(self):
        print(f"""Player Name: {self.name}
Player Season Earning without bonus: {self.earning}
Bonus: {self.bonus}
Player Season Earning After Bonus: {self.earning+self.bonus}
""")

 
print("**********************")
player1 = PlayerEarning('Buffon')
player1.calculateTotal(250000)
player1.printDetails()
print("\n**********************")
player2 = PlayerEarning('Dybala')
player2.calculateTotal(250000,31)
player2.printDetails()
print("\n**********************")
player3 =PlayerEarning('Cuadrado')
player3.calculateTotal(250000,20)
player3.printDetails()  
```

```python
class Player:
    database = {}
    playerNo = 0
    def __init__(self, name, team, jerseyNo):
        self.name = name
        self.team = team
        self.jerseyNo = jerseyNo
    def __str__(self):
        return "Name:{}\nTeam:{}\nJersey No:{}".format(self.name,self.team,self.jerseyNo)

class FootballPlayer(Player):
    
    def __init__(self,name,team,jerseyNo,goals,retire="Not yet retired"):
        super().__init__(name,team,jerseyNo)
        self.goals=goals
        self.retire_stat=retire
        Player.playerNo+=1
        self.player_id= str(Player.playerNo)+self.name.split(" ")[0][0]+self.name.split(" ")[1][0]+str(self.jerseyNo)

        Player.database[self.player_id]=[self.name,self.team,self.jerseyNo,self.goals,self.retire_stat]
    @classmethod
    def createPlayer(cls,name,team,jerseyNo,goals,retirement):
        return cls(name,team,jerseyNo,goals,retirement)
   
    def __str__(self):
        x=f"Player ID: {self.player_id}\n"
        x+=super().__str__()+f"Goals Scored:{self.goals}\nRetirement date:{self.retire_stat}"
        return x  

print("Number of players:", Player.playerNo)
print("Player Database:", Player.database)
print("#################################")
p1 = FootballPlayer("Lionel Messi","Barcelona",10,231)
print("------Details of the player------")
print(p1)
print("#################################")
p2 = FootballPlayer("Cristiano Ronaldo","Juventus",7,215)
print("------Details of the player------")
print(p2)
print("#################################")
p3 = FootballPlayer.createPlayer("Miroslav Klose","Lazio",11,71,"11 Aug,2014")
print("------Details of the player------")
print(p3)
print("#################################")
print("Number of players:",Player.playerNo)
print("Player Database:",Player.database)


```

```python
class Library:
    Total_book = 1000
    borrow_data = {}

    def __init__(self,n,id):
        self.student_name = n
        self.student_id = id
    def borrowbook(self):
        print("A book is borrowed!")
    def __str__(self):
        return "Library: XYZ"

class Student(Library):
    def borrowbook(self,name,uid=""):
        super().borrowbook()
        Library.Total_book-=1
        uid=uid
        bname=name
        self.borrowed_list=[]
        self.borrowed_list.append(bname)
        self.statement=','.join(self.borrowed_list)
        if bname not in Library.borrow_data:
            Library.borrow_data[bname]=[self.student_name]
        print(f"{bname} book with the unique id {uid} is borrowed by {self.student_name}({self.student_id})")
        print(f"Number of books availabe for borrowing = {Library.Total_book}")
        
    def __str__(self):
        
        x= super().__str__() + f"Student Name: {self.student_name} ID: {self.student_id}\nBooks borrowed: {self.statement} "
 
        return x
    def returnAllBooks(self):
        del_books = [book for book, user in Library.borrow_data.items() if self.student_name in user]
        for books in del_books:
            del Library.borrow_data[books]
        
s1 = Student("Alice",18101259)
s1.borrowbook("The Alchemist", "Hdw652")
print("===============")
print(s1)
print("===============")
print(Library.borrow_data)
print("===============")
s1.borrowbook("Wuthering Heights")
print("===============")
print(s1)
print("===============")
s2= Student("David",18141777)
s2.borrowbook("The Alchemist", "Hdw652")
print("===============")
s2.borrowbook("The Vampyre")
print("===============")
print(Library.borrow_data)
print("===============")
s1.returnAllBooks()
print("===============")
print(Library.borrow_data)       
```

```python
class User:
    
    def __init__(self,name,destination,ride_type="Single"):
        self.name=name
        self.destination=destination
        self.type=ride_type
        self.status_statement= "Status: {self.name} is looking for a {self.type}!"

    
    def status(self):
        print(self.status_statement)


class Uber(User):
    
    def __init__(self,car_num,type,*routes):
        self.car_number=car_num
        self.service_type=type
        self.routelist=list(routes)
        self.routes='-->'.join(routes)
        
    def details(self):
        print(f"Car number: {self.car_number}\nType: {self.service_type}\nRoutes: {self.routes}")
        
        
    def pick(self,*user_objs):
        for obj in user_objs:
            if obj.type == self.service_type:
                if obj.destination in self.routelist:
                    obj.status_statement = f"Status: {obj.name} boarded in car {self.car_number}!"
                    print(f"{obj.name} has been picked up.")
                else:
                    
                    print(f"{obj.name}'s destination is different from this car's route.")
            else:
                print(f"{obj.name} is looking for a different ride.")
                
                
user1 = User("Brooks", "Banani", "Shared")
user2 = User("Jocelyn", "Uttara")
user3 = User("Robert", "Gulshan", "Shared")
user4 = User("Langdon", "Mohakhali","Shared")
user1.status()
user2.status()
user3.status()
user4.status()
print("----------------------------------")
car1 = Uber("0K32BH", "Shared",
"Mohakhali", "Banani", "Nikunja", "Uttara")
car1.details()
print("----------------------------------")
car1.pick(user1,user2,user3,user4)
print("----------------------------------")
user1.status()
user2.status()
user3.status()
user4.status()
print("----------------------------------")
car2 = Uber("5GD2BD", "Single", "Uttara")
car3 = Uber("4T12FR", "Shared", "Gulshan","Bashundhara")
car2.details()
car3.details()
print("----------------------------------")
car2.pick(user2, user3)
print("----------------------------------")
car3.pick(user3)
print("----------------------------------")
user2.status()
user3.status()
        
```
