Dot product

```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# Element-wise addition
result_addition = a + b

# Element-wise multiplication
result_multiplication = a * b

# Dot product
dot_product = np.dot(a, b)

print(result_addition)
print(result_multiplication)
print(dot_product)
```

```python
a = [1, 2, 3]
b = [4, 5, 6]

# Manual computation of dot product using only loops
dot_product = 0
for i in range(len(a)):
    # In the first iteration (i = 0):
    #   dot_product += a[0] * b[0]  =>  dot_product += 1 * 4  =>  dot_product = 4  

    # In the second iteration (i = 1):
    #   dot_product += a[1] * b[1]  =>  dot_product += 2 * 5  =>  dot_product = 14

    # In the third iteration (i = 2):
    #   dot_product += a[2] * b[2]  =>  dot_product += 3 * 6  =>  dot_product = 32
    dot_product += a[i] * b[i]

print(dot_product)
```
