## Difference 2: Adding new element

```python
list1 = [2,4,5]
list1.append(7)
print(list1)
```

```python
array1 = np.array([2,4,5])
array1.append(7)
print(array1)
```

```python
array2 = np.zeros(shape = (4))

for i in range(len(array1)): #[2 4 5 0]

   array2[i] = array1[i]


array2[3] = 7    #[2  4  5   7]

print(array2)
```

## Adding list vs array

```python
l1 = [2,3]
l2 = [5,6]

l1 + l2
```

```python
a1 = np.array([2,3])
a2 = np.array([5,6])

print(a1+a2)
```

```python
final = np.zeros(shape = len(a1))


for i in range(len(a1)):

   final[i] = a1[i] + a2[i]

print(final)
```

## Element Addition

```python
sum = 0

for i in range(len(a1)):

   sum += a1[i]

print(sum)
```
