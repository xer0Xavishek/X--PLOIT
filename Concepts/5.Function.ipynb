````python
#Here the number of parameter = Number of arguments

def summation(n1,n2): #two parameters received
   print(n1+n2)


summation(3,4) #two arguments passed
````

````python
def summation(n1,n2):#But two parameter are waiting to reveive value. Since mismatch so there is an ERROR
   print(n1+n2)


summation(3)  #One argument passed
````

````python
def summation(n1,n2=4):#               Two parameter. First one recieves a value. Second parameter is default.
   print(n1+n2)


summation(3)#One argument passed  - default paramter is used. Second prarameter does not receive any value
summation(3,6)#Two argument passed - default parameter gets replaced. Second prarameter receives value
````

````python
def summation(*args):#Two parameter. First one recieves a value. Second parameter is default.
  print(args,len(args))
  if len(args) == 1:

   print(args[0]+4)

  elif len(args) == 2:
    print(args[0]+args[1])



summation(3)
summation(3,6)
summation(3,6,7,8)
````

````python
def summation(*args):#Two parameter. First one recieves a value. Second parameter is default.

 sum =0
 for i in args:
   sum += i

 print(sum)

summation(3)
summation(3,6)
summation(3,6,7,8)
````







`If default parameter does not receive a value then it will hold the default value.
If default paramter receives a value, then it will hold the received value.`

````python
def highestScorer(players, scores):
    if len(players) != len(scores):
        return "Error: lists must be of equal length"
    
    highest_score = max(scores)
    highest_scorer_index = scores.index(highest_score)
    highest_scorer = players[highest_scorer_index]
    
    print(f"The highest scorer was {highest_scorer} with a score of {highest_score}")

# Test the function
players = ["Player1", "Player2", "Player3"]
scores = [100, 200, 150]
highestScorer(players, scores)
````
