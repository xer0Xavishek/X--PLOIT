```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function for the differential equation
def f(x, y, yp, ypp):
    return ypp + yp - y - ypp

# Euler's method
def euler_method(f, x0, y0, yp0, ypp0, xf, h):
    num_steps = int((xf - x0) / h) + 1
    x_values = np.linspace(x0, xf, num_steps)
    y_values = np.zeros(num_steps)
    yp_values = np.zeros(num_steps)
    ypp_values = np.zeros(num_steps)

    y_values[0] = y0
    yp_values[0] = yp0
    ypp_values[0] = ypp0

    for i in range(1, num_steps):
        ypp = f(x_values[i-1], y_values[i-1], yp_values[i-1], ypp_values[i-1])
        yp = yp_values[i-1] + h * ypp
        y = y_values[i-1] + h * yp

        y_values[i] = y
        yp_values[i] = yp
        ypp_values[i] = ypp

    return x_values, y_values

# Initial conditions
x0 = 0
y0 = 7
yp0 = -3
ypp0 = -1

# Step size
h = 0.01

# Final x value
xf = 5

# Solve using Euler's method
x_values, y_values = euler_method(f, x0, y0, yp0, ypp0, xf, h)

# Plot
plt.plot(x_values, y_values)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Solution using Euler\'s method')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


def diff_eq(x, y):
    return y[1], y[2], y[2] - y[1] - y[0]

def euler_method(diff_eq, initial_conditions, x_range, step_size):
    x_values = np.arange(x_range[0], x_range[1] + step_size, step_size)
    y_values = np.zeros((len(x_values), len(initial_conditions)))
    y_values[0] = initial_conditions
    
    for i in range(1, len(x_values)):
        dydx = diff_eq(x_values[i-1], y_values[i-1])
        y_values[i] = y_values[i-1] + step_size * np.array(dydx)
    
    return x_values, y_values


initial_conditions = [7, -3, -1]

x_range = (0, 2)
step_size = 0.01

x_values, y_values = euler_method(diff_eq, initial_conditions, x_range, step_size)

plt.plot(x_values, y_values[:, 0])
plt.xlabel('x')
plt.ylabel('y')
plt.title("Solution using Euler's method")
plt.grid(True)
plt.show()
```

```python
#bb
from sympy import symbols, Function, dsolve

# Define symbols
x = symbols('x')
y = Function('y')(x)

# Define the differential equation
diff_eq_sympy = y.diff(x, 3) + y.diff(x, 2) - y.diff(x) - y

# Solve the differential equation
solution = dsolve(diff_eq_sympy, y, ics={y.subs(x, 0): 7, y.diff(x).subs(x, 0): -3, y.diff(x, 2).subs(x, 0): -1})

# Evaluate the solution at x = 1.5
y_value_at_1_5 = solution.subs(x, 1.5).rhs.evalf()

print("Value of y at x = 1.5:", round(y_value_at_1_5, 2))
```

```python
#22 AAA 
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return 2*x**3 - 5*x**2 + 2*x + 1

# Define the derivative of the function
def f_prime(x):
    return 6*x**2 - 10*x + 2

# Find critical points by solving f'(x) = 0
critical_points = np.roots([6, -10, 2])

# Use the second derivative test to classify critical points
maxima = [point for point in critical_points if f_prime(point) < 0]
minima = [point for point in critical_points if f_prime(point) > 0]

# Generate x values for plotting
x_values = np.linspace(-5, 5, 400)
y_values = f(x_values)

# Plot the function
plt.plot(x_values, y_values, label='f(x) = 2x^3 - 5x^2 + 2x + 1')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of function with maxima and minima')
plt.grid(True)

# Mark maxima and minima with star markers
plt.scatter(maxima, f(np.array(maxima)), color='red', marker='*', label='Maxima')
plt.scatter(minima, f(np.array(minima)), color='green', marker='*', label='Minima')

plt.legend()
plt.show()
```

```python
# 2bbbbb
# Define the function
def g(x):
    return np.sin(2*x) + np.cos(3*x)

# Define the derivative of the function
def g_prime(x):
    return 2*np.cos(2*x) - 3*np.sin(3*x)

# Find critical points by solving f'(x) = 0
critical_points_g = np.roots([-3, 0, -2])

# Use the second derivative test to classify critical points
maxima_g = [point for point in critical_points_g if g_prime(point) < 0]
minima_g = [point for point in critical_points_g if g_prime(point) > 0]

# Generate x values for plotting
x_values_g = np.linspace(0, 3*np.pi/4, 400)
y_values_g = g(x_values_g)

# Plot the function
plt.plot(x_values_g, y_values_g, label='f(x) = sin(2x) + cos(3x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of function with maxima and minima')
plt.grid(True)

# Mark maxima and minima with star markers
plt.scatter(maxima_g, g(np.array(maxima_g)), color='red', marker='*', label='Maxima')
plt.scatter(minima_g, g(np.array(minima_g)), color='green', marker='*', label='Minima')

plt.legend()
plt.show()
```

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Define the system of equations
def system(y, t):
    R, W = y
    dR_dt = R * (0.08 - 0.001 * W)
    dW_dt = W * (-0.03 - 0.00002 * R)
    return [dR_dt, dW_dt]

# Initial conditions
R_0 = 100
W_0 = 40
initial_conditions = [R_0, W_0]

# Create a time array for plotting
time_array = np.linspace(0, 100, 1000)

# Solve the system of equations
solution = odeint(system, initial_conditions, time_array)

# Extract the solutions
R_values, W_values = solution.T

# Plotting W vs t
plt.figure(figsize=(10, 5))
plt.plot(time_array, W_values, label='Wolves')
plt.title('Population of Wolves Over Time')
plt.xlabel('Time')
plt.ylabel('Population')
plt.grid(True)
plt.legend()
plt.show()

# Plotting W vs R
plt.figure(figsize=(8, 6))
plt.plot(R_values, W_values)
plt.title('Phase Portrait: Wolves vs Rabbits')
plt.xlabel('Rabbits')
plt.ylabel('Wolves')
plt.grid(True)
plt.show()
```
