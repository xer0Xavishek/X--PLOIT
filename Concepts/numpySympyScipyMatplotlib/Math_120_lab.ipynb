# Rectangular Method(Mid Point Rule)

$\int_{0}^{1}\frac{1}{1+x^2}dx$

```python
def f(x):
  return 1/(1+x**2)

a=0
b=1
n=1000
delx=(b-a)/n

sum=0
for i in range(n):
  # Find the mid point for each rectangle
  x_mid=a+delx*(i+0.5)
  sum+=f(x_mid)
print(delx*sum)
```

# Trapezoidal Rule

$\int_{1}^{2}x^2dx$

```python
def f(x):
  return x**2

x0=1# lower lim
xn=2# upper lim
n=100# num of Trap
delx=(xn-x0)/n # width of each trapezoid

sum=f(x0)+f(xn) # convention to add first and last funcnal val
for i in range(1,n):
  x=x0+delx*i
  sum+=2*f(x)
print(0.5*delx*sum)
```

# Simpson's 1/3 Rule

$\int_{1}^{2}x^2dx$

```python
def f(x):
  return x**2

x0=1
xn=2
n=4
delx=(xn-x0)/n

sum=f(x0)+f(xn)
for i in range(1,n):
  x=x0+delx*i
  if i%2!=0:
   sum+=4*f(x)
  elif i%2==0:
    sum+=2*f(x)
print(delx/3*sum)
```

# $\int_{0}^{\pi}sin(x)$

```python
import numpy as np
import matplotlib.pyplot as plt

f = lambda x: np.sin(x)
a = 0
b = np.pi
N = 1000

plt_vals = []
for i in range(N):
    x = np.random.uniform(low=a, high=b, size=N)
    sum = 0
    for i in x:
        sum += f(i)
    res = (b - a) * sum / N
    plt_vals.append(res)
plt.hist(plt_vals, bins=20, ec="black")
plt.xlabel("Areas")
plt.ylabel("Frequency")
plt.show()
```

```python
# Find the first derivative
diff1=sp.diff(f,x)
# Find the second derivative
diff2=sp.diff(f,x,2)

# Get the solution of the first derivative
soln=sp.solve(diff1,x)

# store the points where maxima minima exist in a list
x_pt=[]
for i in range(len(soln)):
  # Test the value of the second derivative at each solution
  # Also test whether the solution is real or imaginary
  if sp.im(soln[i])==0 and diff2.subs(x,soln[i])>0:
    x_pt.append(soln[i])
    # find the minimum value of the function
    min_val=f.subs(x,soln[i])
    print(f'The minimum value of the function at {soln[i]} is {min_val}')
  elif sp.im(soln[i])==0 and diff2.subs(x,soln[i])<0:
    x_pt.append(soln[i])
    # find the maximum value of the function
    max_val=f.subs(x,soln[i])
    print(f'The maximum value of the function at {soln[i]} is {max_val}')
  else:
    continue
```

```python
marker=[]
for i in range(len(x_pt)):
  # find the x,y coordinate for the maximum and minimum points
  X=x_pt[i]
  Y=f.subs(x,x_pt[i])
  # Add a r* marker in the marker list
  marker.append({'args':[X,Y,'r*']})
# Plot the function with the marker list
sp.plot(f,(x,-4,4),markers=marker)
```

# System of differential equation

image.png

The problem can be modelled as a system of differential equation.

Apply Newton's second law on $m_1$:

$F=m_1a_1$

The sum of forces on $m_1$ are composed of-

1) Elastic force exerted by $k_1$ spring:

$-k_1x_1$

2) Elastic force exerted by $k_2$ spring:

$-k_2(x_1-x_2)$

3) $20N$ force applied in the positive direction.

So, the Newton's law can be written as:

$-k_1x_1-k_2(x_1-x_2)+20=m_1a_1$

Rewriting the equation in differtial form:

$-k_1x_1-k_2(x_1-x_2)+20=m_1d^2x_1/dt^2$---(1)

Similarly, applying Newton's second law on $m_2$ gives us:

$-k_3x_2-k_2(x_2-x_1)=m_2d^2x_2/dt^2$------(2)


```python
# Define system params
m1=2
m2=9/4
k1=1
k2=3
k3=15/4
F=20
# Declare the symbol t
t=sp.symbols('t')

# Write x1 as a function of t
x1=sp.Function('x1')(t)
# Write x2 as a function of t
x2=sp.Function('x2')(t)

# Write the first differential equation
Eq1=sp.Eq(m1*sp.diff(x1,t,2),-k1*(x1)-k2*(x1-x2)+F)
display(Eq1)
# Write the second differential equation
Eq2=sp.Eq(m2*sp.diff(x2,t,2),-k3*x2-k2*(x2-x1))
display(Eq2)
# Take the equations in a list to make it a system
system=[Eq1, Eq2]

# define the initial conditions in a dictionary
boundary={x1.subs(t,0):0,sp.diff(x1,t).subs(t,0):0,x2.subs(t,0):0,sp.diff(x2,t).subs(t,0):0}

# solve the system using dsolve
sol=sp.dsolve(system,ics=boundary)
display(sol)
```

```python
# Take the plot of the first solution x1(t) in p1
p1=sp.plot(sol[0].rhs, (t,0,10),show=False)
# Take the plot the second solution x2(t) in p2
p2=sp.plot(sol[1].rhs, (t,0,10),show=False)
# Extend p1 plot up to p2
p1.extend(p2)
# show the p1 plot
p1.show()
```

# Euler method of solving differential equation

$dy/dx=y$

$y(0)=1$

```python
import numpy as np
from matplotlib.pyplot import *
# lower boundary
a=0
# Upper boundary
b=5
# Number of point within the boundary
N=1001
# distance between each point
h=(b-a)/(N-1)
# N equidistant points in x direction
x=np.linspace(a,b,N)
# y is an array of zeros. It's size is equal to x
y=x*0
# intial value of y
y[0]=1
# estimating the value of y at each point of x except the intial value which is given at x=0
for i in range(1,N):
  y[i]=y[i-1]+h*y[i-1]

plot(x,y)
grid()
show()

```

```python
f=lambda x,t:np.sin(3*t)-16*x
a=0
b=3
N=10001
h=(b-a)/(N-1)
t=np.linspace(a,b,N)
x=t*0
v=x*0
v[0]=0
x[0]=2
for i in range(1,N):
  v[i]=v[i-1]+h*f(x[i-1],t[i-1]) # First order estimation
  x[i]=x[i-1]+h*v[i] # Second order estimation
plot(t,x)
grid()
show()
```

# Runge-kutta 4th Order
https://en.wikipedia.org/wiki/Runge%E2%80%93Kutta_methods

image.png

$dy/dx=y$

$y(0)=1$

```python
import numpy as np
import matplotlib.pyplot as plt
a=0
b=2
N=6
h=(b-a)/(N-1)
x=np.linspace(a,b,N)
y=0*x
y[0]=1
for i in range(1,N):
  # slope at previous point
  k1=y[i-1]
  # Slope at midpoint using k1
  k2=y[i-1]+h/2*k1
  # Slope at midpoint using k2
  k3=y[i-1]+h/2*k2
  # Slope at the point to estimation using k3
  k4=y[i-1]+h*k3
  y[i]=y[i-1]+h/6*(k1+2*k2+2*k3+k4)
plt.plot(x,y)
plt.plot(np.linspace(0,2,200),np.exp(np.linspace(0,2,200)))
```

Look at the sharp difference between between Euler method and the Range-Kutta 4th order. Both use $6$ equidistant points within the boundary 1 to 2 to solve ode $dy/dx=y$.

However, the Euler method solution hugely differs from the actual analytical solution which is $e^x$ compared to Range-Kutta 4th order.

```python
import numpy as np
import matplotlib.pyplot as plt
a=0
b=2
N=6
h=(b-a)/(N-1)
x=np.linspace(a,b,N)
y=0*x
y[0]=1
for i in range(1,N):
  y[i]=y[i-1]+h*y[i-1]
plt.plot(x,y)
plt.plot(np.linspace(0,2,200),np.exp(np.linspace(0,2,200)))
```
